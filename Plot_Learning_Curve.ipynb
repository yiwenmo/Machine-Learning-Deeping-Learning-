```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#read data
train = pd.read_csv('training.csv',sep = ",")

acc_1 = train['accuracy']
val_acc_1 = train['val_accuracy']
loss_1 = train['loss']
val_loss_1 = train['val_loss']
epochs = train['epoch']

#plot loss curve
plt.plot(epochs, loss_1, 'b', label='loss_1')
plt.plot(epochs, val_loss_1, 'g', label='val_loss_1')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.savefig('problem2_1_loss.png')
plt.show()
```

```python
#plot accuracy curve
plt.plot(epochs, acc_1, 'b', label='acc_1')
plt.plot(epochs, val_acc_1, 'g', label='val_acc_1')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('accuracy')
plt.legend()
plt.savefig('problem2_1_acc.png')
plt.show()
```
