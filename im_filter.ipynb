```python
%matplotlib inline
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pylab as plt
im = mpimg.imread("lenna-gray.tif") 
```

```python
print(im.shape,type(im))
plt.imshow(im,cmap = 'gray')
```

```python
import numpy as np
from scipy.signal import convolve2d as conv2
kernel = np.array([[0,1,0],[1,-4,1],[0,1,0]]) #laplacian filter
kernel
output = conv2(im, np.flip(kernel), mode ='valid')
plt.imshow(output,cmap = 'gray')
```
