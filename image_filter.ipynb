```python
%matplotlib inline
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pylab as plt
import numpy as np
from scipy.signal import convolve2d as conv2
```

```python
im = mpimg.imread("cat1.jpg") 
print(im.shape,type(im)) #rgb_image
plt.imshow(im)
im = im[:,:,1] #turn to 2D image
print(im.shape,type(im))
plt.savefig('sobel_original.png')
```

```python
import numpy as np
from scipy.signal import convolve2d as conv2
#sobel y-axis filter
kernel = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
kernel
output = conv2(im, np.flip(kernel), mode ='valid')
plt.imshow(output,cmap = 'gray')
plt.savefig('sobel.png')
```

```python
im = mpimg.imread("cat3.jpg") 
print(im.shape,type(im)) #rgb_image
plt.imshow(im)
im = im[:,:,1] #turn to 2D image
print(im.shape,type(im))
plt.savefig('blur_original.png')
```

```python
#blur filter
kernel = np.ones((10,10), dtype = 'float') * (1.0/(10*10))
kernel
output = conv2(im, np.flip(kernel), mode ='valid')
plt.imshow(output,cmap = 'gray')
plt.savefig('blur_.png')
```

```python
im = mpimg.imread("dog.jpg") 
print(im.shape,type(im)) #rgb_image
plt.imshow(im)
plt.savefig('sharpen_original.png')
im = im[:,:,1] #turn to 2D image
print(im.shape,type(im))
```

```python
#sobel x filter
kernel = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
kernel
output = conv2(im, np.flip(kernel), mode ='valid')
plt.imshow(output, cmap = 'gray')
plt.savefig('sobelx.png')
```
